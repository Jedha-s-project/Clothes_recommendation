```python
datatest = pd.read_excel("./Dataset/datatest.xlsx")
datatest
```

```python
datatest["stsb-xlm-r-multilingual_cos_sim"] = ""
```

```python
datatest = datatest.rename(columns={"mood ": "mood", "clothes ": "clothes"})
```

```python
for i in range (len(datatest)):
    x = model.encode(datatest.loc[i,"mood"])
    y = model.encode(datatest.loc[i,"clothes"])
    X = x.reshape(1, -1)
    Y = y.reshape(1, -1)
    datatest.loc[i,"stsb-xlm-r-multilingual_cos_sim"] = sklearn.metrics.pairwise.cosine_similarity(X,Y)
datatest
```

```python
datatest["stsb-xlm-r-multilingual_cos_sim_transf"] =""
```

```python
for i in range (len(datatest)):
    y_pred = datatest.loc[i,"stsb-xlm-r-multilingual_cos_sim"]
    datatest.loc[i, "stsb-xlm-r-multilingual_cos_sim_transf"] = y_pred[0]*10
datatest
```

```python
y_true = datatest["fc_similarity"]
y_pred = datatest["stsb-xlm-r-multilingual_cos_sim_transf"]
MSE1 = sklearn.metrics.mean_squared_error(y_true,y_pred)
MSE1
```

```python
datatest = datatest.drop(["stsb-xlm-r-multilingual_cos_sim"], axis=1)
datatest
```
