```python
# Import of librairies
import pandas as pd
```

```python
# Import of data
df_clothes = pd.read_excel("./Dataset/Clothes_table.xlsx")
df_clothes
```

```python
df_mood = pd.read_excel("./Dataset/Mood_table.xlsx")
```

```python
df_mood = df_mood.drop(["Unnamed: 0"], axis=1)
df_mood
```

```python
import sklearn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("stsb-xlm-r-multilingual")
```

```python
df_clothes["sentence_transformer"] = ""
df_mood["sentence_transformer"] = ""
```

```python
for i in range (len(df_mood)):
    df_mood.loc[i]["sentence_transformer"] = model.encode(df_mood.loc[i]["day_mood"])
```

```python
for i in range (len(df_clothes)):
    df_clothes.loc[i]["sentence_transformer"] = model.encode(df_clothes.loc[i]["description"])
df_clothes
```

```python
df_mood["cos_sim_blouse_1"]=""
```

```python
for i in range (len(df_mood)) :
    X = df_clothes.loc[8]["sentence_transformer"].reshape(1, -1)
    Y = df_mood.loc[i]["sentence_transformer"].reshape(1, -1)
    df_mood.loc[i]["cos_sim_blouse_1"] = sklearn.metrics.pairwise.cosine_similarity(X,Y)
```
