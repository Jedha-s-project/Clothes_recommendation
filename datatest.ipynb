```python
# Import of librairies
import pandas as pd
```

```python
# Import of data
datatest = pd.read_excel("Clothes_bdd.xlsx", header=[1])
```

```python
# Delete first column
datatest = datatest.iloc[:, 1:]
```

```python
# Columns rename
datatest = datatest.rename(columns={"Noms":"name", "Photos" :"photos", "Description" : "descriptions"})
```

```python
# New column for category
datatest['category'] = ""
```

```python
# category from name
for i in range (len(datatest)) :
    x = datatest.iloc[i][0].split("_")
    datatest.loc[i]['category']= x[0].lower()
datatest
```
