```python
# Import of librairies
import pandas as pd
```

```python
# Import of data
df_clothes = pd.read_excel("Clothes_table.xlsx", header=[1])
```

```python
# Delete first column
df_clothes = df_clothes.iloc[:, 1:]
```

```python
# Columns rename
df_clothes = df_clothes.rename(columns={"Noms":"name", "Photos" :"photos", "Description" : "descriptions"})
```

```python
# New column for category
df_clothes['category'] = ""
```

```python
# Category from name
for i in range (len(df_clothes)) :
    x = df_clothes.iloc[i][0].split("_")
    df_clothes.loc[i]['category']= x[0].lower()
df_clothes
```

```python
df_mood = pd.read_excel("Mood_table.xlsx")
df_mood
```

```python
df_mood = df_mood.drop(["Unnamed: 0"], axis=1)
df_mood
```

```python
import fr_core_news_sm
nlp = fr_core_news_sm.load()
```

```python
from spacy.lang.fr.stop_words import STOP_WORDS
```

```python
# Get all string punctuation that we want to remove. 
import string 
string.punctuation
```

```python
df_clothes['clean_descriptions'] = df_clothes['descriptions'].str.replace(r"[!\"#$%&()*+,-./:;<=>?@[\\\]^_`{|}~]+", " ") 
df_mood['clean_day_mood'] = df_mood['day_mood'].str.replace(r"[!\"#$%&()*+,-./:;<=>?@[\\\]^_`{|}~]+", " ") 
df_clothes['clean_descriptions'] = df_clothes['clean_descriptions'].fillna('').apply(lambda x: x.lower())
df_mood['clean_day_mood'] = df_mood['clean_day_mood'].fillna('').apply(lambda x: x.lower())
df_mood
```

```python
df_clothes
```

```python
## Tokenize the cleaned document
tokenized_clothes = df_clothes['clean_descriptions'].fillna('').apply(lambda x: nlp(x))
tokenized_mood = df_mood['clean_day_mood'].fillna('').apply(lambda x: nlp(x))
```

```python
# remove stop-words
tokenized_clothes = tokenized_clothes.apply(lambda x: [token.lemma_ for token in x if token.text not in STOP_WORDS])
tokenized_mood = tokenized_mood.apply(lambda x: [token.lemma_ for token in x if token.text not in STOP_WORDS])
```

```python
tokenized_clothes
```

```python
tokenized_mood
```

```python
# add a clean token column 
df_clothes["clean_token"] = [" ".join(x) for x in tokenized_clothes]
df_mood["clean_token"] = [" ".join(x) for x in tokenized_mood]
```

```python
df_clothes
```

```python
df_mood
```

```python
# Tfidf transformer 
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF vector
vectorizer_clothes = TfidfVectorizer()
X_clothes = vectorizer_clothes.fit_transform(df_clothes['clean_token'])

# X is a generator. We can transform that as an array
dense_clothes = X_clothes.toarray()
dense_clothes
```

```python
vectorizer_clothes.vocabulary_
```

```python
# TF-IDF vector
vectorizer_mood = TfidfVectorizer()
X_mood = vectorizer_mood.fit_transform(df_mood['clean_token'])

# X is a generator. We can transform that as an array
dense_mood = X_mood.toarray()
dense_mood
```

```python
vectorizer_mood.vocabulary_
```

```python
clothes_tfidf_matrix = pd.DataFrame(dense_clothes, 
                            columns=[x for x in vectorizer_clothes.get_feature_names()],
                            index = ["desc_clothes_{}".format(x) for x in df_clothes["name"]])
```

```python
clothes_tfidf_matrix
```

```python
mood_tfidf_matrix = pd.DataFrame(dense_mood, 
                            columns=[x for x in vectorizer_mood.get_feature_names()],
                            index=["desc_mood_{}".format(x) for x in range(len(df_mood))])
```
