```python
import torch
```

```python
from transformers import CamembertModel, CamembertTokenizer
from transformers import CamembertConfig
# (Need to reload the model with new config)
```

```python
tokenizer = CamembertTokenizer.from_pretrained('camembert-base')
camembert = CamembertModel.from_pretrained('camembert-base')
```

```python
df_mood['tokenized_mood'] = ""
```

```python
df_mood.loc[0]['tokenized_mood']=tokenizer.tokenize(df_mood.loc[0]['day_mood'])
```

```python
for i in range (len(df_mood)):
    df_mood.loc[i]['tokenized_mood']=tokenizer.tokenize(df_mood.loc[i]['day_mood'])
```

```python
df_mood['encoded_mood'] = ""
```

```python
for i in range (len(df_mood)):
    df_mood.loc[i]['encoded_mood']=tokenizer.encode(df_mood.loc[i]['tokenized_mood'])
df_mood
```

```python
df_clothes['tokenized_clothes'] = ""
df_clothes = df_clothes.drop(["Unnamed: 0"], axis=1)
```

```python
df_clothes.loc[0]['tokenized_clothes']=tokenizer.tokenize(df_clothes.loc[0]['descriptions'])
```

```python
for i in range (len(df_clothes)):
    df_clothes.loc[i]['tokenized_clothes']=tokenizer.tokenize(df_clothes.loc[i]['descriptions'])
```

```python
df_clothes['encoded_clothes'] = ""
```

```python
for i in range (len(df_clothes)):
    df_clothes.loc[i]['encoded_clothes']=tokenizer.encode(df_clothes.loc[i]['tokenized_clothes'])
df_clothes
```

```python
encoded_sentence = torch.tensor(df_clothes.loc[0]['encoded_clothes']).unsqueeze(0)
encoded_sentence
```

```python
embeddings= camembert(encoded_sentence,output_hidden_states=True)
embeddings
```

```python
df_clothes["CLS"] =""
df_clothes
```

```python
for i in range (len(df_clothes)):
    encoded_sentence = torch.tensor(df_clothes.loc[i]['encoded_clothes']).unsqueeze(0)
    embeddings= camembert(encoded_sentence,output_hidden_states=True)
    last_hidden_states = embeddings.hidden_states[-1]
    df_clothes.loc[i]["CLS"] = last_hidden_states[0,0,:]
```

```python
for i in range (len(df_mood)):
    encoded_sentence = torch.tensor(df_mood.loc[i]['encoded_mood']).unsqueeze(0)
    embeddings= camembert(encoded_sentence,output_hidden_states=True)
    last_hidden_states = embeddings.hidden_states[-1]
    df_mood.loc[i]["CLS"] = last_hidden_states[0,0,:]
```
